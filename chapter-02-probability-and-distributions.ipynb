## üé≤ Probability Basics

- **Definition:**  
  $$
  \text{Probability} = \frac{\text{\# of ways an event can happen}}{\text{total \# of possible outcomes}}
  $$
- **Examples:**

  - Coin flip:  
    $$
    P(\text{Heads}) = \frac{1}{2} = 50\%
    $$
  - Assigning people randomly:  
    $$
    P(\text{Brian}) = \frac{1}{4} = 25\%
    $$

## üîó Independent vs Dependent Events

- **Independent Events:** One event doesn‚Äôt affect the next  
  _(e.g., flipping a coin twice)_
  
- **Dependent Events:** One outcome affects another  
  _(e.g., drawing names without replacement)_

## üîç Conditional Probability

- **Formula:**  
  $$
  P(A|B) = \frac{P(A \cap B)}{P(B)}
  $$

### üßæ Example: What is the chance of a kitchen order being over $150?

- Total kitchen orders = **181**  
- Kitchen orders over $150 = **20**

$$
P(\text{Order} > 150 \mid \text{Kitchen}) = \frac{20}{181} \approx 11.05\%
$$

### üßæ Example: What is the chance that an order over $150 will be a kitchen product?

- Orders over $150 = **601**  
- Kitchen orders over $150 = **20**

$$
P(\text{Kitchen} \mid \text{Order} > 150) = \frac{20}{601} \approx 3.33\%
$$

## üé≤ Probability Distributions

### üì¶ What is a Probability Distribution?

A function that shows the likelihood of all possible outcomes.

**Example:** Rolling a six-sided die:  
Each outcome (1‚Äì6) has a probability of  
$$
\frac{1}{6}
$$  
That‚Äôs a **discrete uniform distribution**‚Äîequal odds for all.

### üìå Why are Probability Distributions Important?

- üìä Quantify uncertainty and risk

- üß™ Critical for hypothesis testing and inference

### üßÆ Expected Value: The Mean of a Probability Distribution

Long-run average result from repeated trials.

**Formula (for a fair die):**
$$
E = \frac{1+2+3+4+5+6}{6} = 3.5
$$

You‚Äôll never roll a 3.5, but over many rolls, your average gets closer and closer to it.

### üéØ Discrete Probability Distributions

Used when outcomes are **countable**.

**Examples:**

- Coin flips
- Dice rolls
- Number of website visitors per day

### üìà Law of Large Numbers
>
> ‚ÄúAs your sample size grows, the sample mean approaches the expected value.‚Äù

**Example:**

- 10 rolls ‚Üí lots of fluctuation  
- 10,000 rolls ‚Üí mean hovers around 3.5

### üåä Continuous Probability Distributions

Used when outcomes can fall **anywhere within a range** (infinite possibilities).

**Examples:**

- Wait time for a bus
- Human height or weight

**You calculate probabilities over a range**, not exact values:
$$
P(4 \leq \text{wait time} \leq 7)
$$

**Key Type:**  

- **Normal distribution** (the bell curve)
